# 메뉴 split

```python
train = pd.read_csv("data/train_join_v3.csv")
train.head()

test = pd.read_csv("data/test_join_v3.csv")
test.head()
```

```python
# Basic text preprocessing

def split_process(x):
    x_ = []
    x = x.split(' ')
    for i in x:
        if '(' in i and ':' in i and ')' in i:
            continue
        if '/' in i:
            x_.extend(i.split('/'))
        elif '*' in i:
            x_.extend(i.split('*'))
        else:
            x_.append(i)
    x_ = list(set(x_))
    x_.remove('')
    return x_
```

```python
train['조식메뉴_split'] = train['조식메뉴'].apply(lambda x: split_process(x)).to_list()
train['중식메뉴_split'] = train['중식메뉴'].apply(lambda x: split_process(x)).to_list()
train['석식메뉴_split'] = train['석식메뉴'].apply(lambda x: split_process(x)).to_list()
```

```python
test['조식메뉴_split'] = test['조식메뉴'].apply(lambda x: split_process(x)).to_list()
test['중식메뉴_split'] = test['중식메뉴'].apply(lambda x: split_process(x)).to_list()
test['석식메뉴_split'] = test['석식메뉴'].apply(lambda x: split_process(x)).to_list()
```
