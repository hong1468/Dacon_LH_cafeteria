```python
pca_df=pd.concat([train_pca,test_pca],ignore_index=True)
```

```python
### PCA

from sklearn.decomposition import PCA
```

```python
pca_df_v2=pca_df.drop(['중식계','석식계'],axis=1)
```

```python
## 다중공선성 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = pca_df_v2.columns
vif['VIF'] = [variance_inflation_factor(pca_df_v2.values, i) for i in range(pca_df_v2.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
list(vif['Features'][:12])
```

```python
# 데이터별 범위가 다르기 때문에 스케일을 맞춰준다.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
nd_scaled = scaler.fit_transform(pca_df_v2)
df_scaled = pd.DataFrame(nd_scaled, columns=pca_df_v2.columns)
df_scaled.describe().round(3)
```

```python
target=df_scaled[['사회적거리두기 1.5단계',
 '격하 가중치',
 '사회적거리두기 1단계',
 '평균기온',
 '최고기온',
 '평균습도',
 '본사정원수',
 '최저기온',
 '최저습도',
 '사회적거리두기 2단계 이상',
 '본사출장자수',
 '격상 가중치']]
```

```python
### PCA
# 일단 대략적으로 6개 columns 로 해보기
num = 6
pca = PCA(n_components=num)
nd_pca = pca.fit_transform(target)

lst_prin = []
for i in range(1,num+1):
    lst_prin.append("Prin" + str(i))

# print(lst_prin)
# 주성분 점수 저장
df_pca = pd.DataFrame(nd_pca, columns=lst_prin)
df_pca.shape
```

```python
# 변수의 크기별 고유값과 누적비율을 비교해서 적용하며 스케일링을 했을 경우에는 고유값 기준을 본다.
# EigenValues > 1 을 기준으로 끊는다.

import numpy as np

df_pca_stat = pd.DataFrame()
df_pca_stat['PrinNo'] = [i for i in range(1,num+1)]
df_pca_stat['EigenValues'] = pca.explained_variance_
df_pca_stat['EigenValueRatio'] = pca.explained_variance_ratio_
df_pca_stat['CumEigenValueRatio'] = np.cumsum(pca.explained_variance_ratio_)
df_pca_stat.round(3)
```

```python
x_train = train_pca_v2[['본사휴가자수', '본사시간외근무명령서승인건수', '현본사소속재택근무자수',
                        '전국 코로나 확진자 수', '경남 코로나 확진자 수', '진주 코로나 확진자 수',
       '1단계 누적일수', '1.5단계 누적일수', '2단계 누적일수', 'Prin1', 'Prin2', 'Prin3',
       'Prin4', 'Prin5', 'Prin6']]
y1_train = train_pca_v2['중식계']
y2_train = train_pca_v2['석식계']

x_test = test_pca_v2[['본사휴가자수', '본사시간외근무명령서승인건수', '현본사소속재택근무자수',
                        '전국 코로나 확진자 수', '경남 코로나 확진자 수', '진주 코로나 확진자 수',
       '1단계 누적일수', '1.5단계 누적일수', '2단계 누적일수', 'Prin1', 'Prin2', 'Prin3',
       'Prin4', 'Prin5', 'Prin6']]
```

```python
model1 = XGBRegressor(colsample_bytree=1, gamma= 0, learning_rate= 0.1, max_depth= 7, min_child_weight= 2, n_estimators= 50,
                       random_state= 2, subsample= 0.5)
model2 = XGBRegressor(colsample_bytree=1, gamma= 0, learning_rate= 0.1, max_depth= 7, min_child_weight= 2, n_estimators= 50,
                       random_state= 2, subsample= 0.5)
```

```python
model1.fit(x_train, y1_train)
model2.fit(x_train, y2_train)
```

```python
submission = pd.read_csv('datasets/sample_submission.csv')
```

```python
pred1 = model1.predict(x_test)
pred2 = model2.predict(x_test)
```

```python
submission['중식계'] = pred1
submission['석식계'] = pred2
```
