# 거리두기 관련 피쳐 추가

```python
for i in range(len(train)):
    if train['일자'][i]<datetime.datetime(2020,8,22):
        train.loc[i,'격상 가중치']=0
        train.loc[i,'격하 가중치']=0
        train.loc[i,'가중치']=0
```

```python
train.head()
```

```python
for i in range(len(train)):
    if train['일자'][i]<datetime.datetime(2020,8,22):
        train.loc[i,'가중치']=0
    else:
        if train['사회적거리두기 2단계 이상'][i]==1:
            train.loc[i,'가중치']=2
        elif train['사회적거리두기 1.5단계'][i]==1:
            train.loc[i,'가중치']=1.5
        elif train['사회적거리두기 1단계'][i]==1:
            train.loc[i,'가중치']=1
```

```python
train['가중치_v2']='.'
train['가중치_v2'][0]=0
for i in range(len(train)-1):
    if train['가중치'][i+1]==train['가중치'][i]:
        train.loc[i+1,'가중치_v2']=train['가중치_v2'][i]
    elif train['가중치'][i+1]!=train['가중치'][i]:
        train.loc[i+1,'가중치_v2']=train['가중치'][i+1]-train['가중치'][i]
```

```python
train[1100:]
```

```python
import math

def dist_inc(x):
    if x==0:
        return 0
    elif x>0:
        return abs(x)
    elif x<0:
        pass
def dist_dec(x):
    if x==0:
        return 0
    elif x<0:
        return abs(x)
    elif x>0:
        pass
```

```python
train['격상 가중치']=train['가중치_v2'].apply(lambda x: dist_inc(x))
train['격하 가중치']=train['가중치_v2'].apply(lambda x: dist_dec(x))
train['격상 가중치']=train['격상 가중치'].fillna(0)
train['격하 가중치']=train['격하 가중치'].fillna(0)
```
